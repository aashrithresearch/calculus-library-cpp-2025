## Polynomial derivatives
Create expressions via `calculus.Expression()`. Structure is: coeff, power, trig (optional). for both polynomials, find f(2) and f'(x)

```python
import calculus
```

```python
expr = calculus.Expression()
```

```python
expr.add_term(3, 2)
expr.add_term(2, 1)
expr.add_term(-5, 0)
```

```python
print(f"f(x) = {expr}")
```

```python
print(f"f(2) = {expr.evaluate(2)}")
```

```python
deriv = calculus.derivative(expr, 1)
```

```python
print(f"f'(x) = {deriv}")
```

```python
deriv_n = calculus.derivative_polynomial(expr, n=2)
```

```python
print(f"f'(x) = {deriv_n}")
```

## Definite Integral

Create expressions via `calculus.Expression()`. Structure is: (coeff, power, trig (optional). for both polynomials, find f(2) and integrate f(x)

```python
expr4 = calculus.Expression()
```

```python
expr4.add_term(10, 0, 'cos')
expr4.add_term(3, 2)
expr4.add_term(-5, 0)
```
